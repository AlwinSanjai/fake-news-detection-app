```python
import joblib

model = joblib.load("logistic_fake_news_model.pkl")
vectorizer = joblib.load("tfidf_vectorizer.pkl")
```

```python
news = ["Government secretly controlling weather"]

news_vector = vectorizer.transform(news)

prediction = model.predict(news_vector)

print(prediction)
```

```python
label_map = {0:"Fake", 1:"True"}

print(label_map[prediction[0]])
```
